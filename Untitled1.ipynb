```python
import pandas as pd
import numpy as np
import requests, time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

# Funciones para selenium
##### Call_driver: inicia la pagina principal
##### cookie_handleer: acepta el cookie del inicio
##### list_maker: hace todo el proceso, llama las dos funciones, scrapea, procesa y cierra el navegador

```python
def call_driver():
    driver = webdriver.Chrome()
    driver.get("https://www.mwcbarcelona.com/exhibitors")
    driver.implicitly_wait(0.5)
    driver.maximize_window()
    return driver
```

```python
def cookie_handler(driver):
    button = driver.find_element(By.CLASS_NAME, value = 'banner-actions-container')
    button.click()
```

```python
def list_maker():
    driver = call_driver()
    cookie_handler(driver)
    
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, '//tr[@role="link"]'))
        )

        element_tr = driver.find_elements(By.XPATH, '//tr[@role="link"]')
        elements_list = [i.get_attribute("data-href") for i in element_tr]
        time.sleep(5)
    except Exception as e:
        print(f"An error occurred: {e}")
        
    driver.quit()
    return elements_list
```

# Funciones para BS

```python
def souper(link_agg):
    url_base = 'https://www.mwcbarcelona.com'
    url_full = url_base + link_agg
    result = requests.get(url_full)
    if result.status_code != 200:
        print(f'Error!!! Respuesta: {result.status_code} - Fallo: {link_agg}')
        try:
            result = requests.get(url_base + link_agg)
            return BeautifulSoup(result.content, 'html.parser')
        except:
            print(f'Fallo!!! Respuesta: {result.status_code} - Fallo: {link_agg}')
            return None
    else:
        return BeautifulSoup(result.content, 'html.parser')
```

```python
def get_info(soup):
    tags_list = []

    exhibitors_ = soup.find('h1')
    information_ = soup.find('p')
    location_ = soup.find('ul', class_='flex flex-wrap')
    tags_ = soup.find_all('li', class_='px-4 py-1 mb-2 mr-2 text-xs no-underline bg-gray-100 border border-gray-300 rounded-full')
    web_page_ = soup.find('a', class_='flex items-center mb-2 hover:opacity-75')
    
    if exhibitors_ is not None:
        a_failure = exhibitors_.get_text().split()[0] != 'Error'
        exhibitors = exhibitors_.get_text() if a_failure else None
    
    if information_ is not None:
        b_fail = information_.get_text().split()[0] != 'NFS' 
        information = information_.get_text() if b_fail else None
        
    location = location_.get_text().strip() if location_ else None
    if tags_:
        for i in tags_:
            tags_list.append(i.get_text().strip())
        tags_array = np.array(tags_list).reshape(-1, 1)
    else:
        tags_array = None

    web_page = web_page_.get('href') if web_page_ else None
    return exhibitors, information, location, tags_array, web_page
```

```python
list_subpages = list_maker()
```

```python
df = pd.DataFrame( columns=['Exhibitors', 'Information', 'Location', 'Tags_list', 'Web_page'])
```

```python
for i in list_subpages:
    souper_result = souper(i)
    if souper_result == None:
        continue
    else:
        df.loc[i] = get_info(souper_result)
```

```python
df = df.reset_index()
df.drop('index', axis = 1, inplace = True)
df
```
